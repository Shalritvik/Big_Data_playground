```python
from pyspark.sql import SQLContext
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
```

```python
from pyspark.sql import SQLContext
sqlsc = SQLContext(sc)
predictions = sqlsc.read.load('/content/drive/My Drive/BigDataColab25FallShalRitvikSinha/predictions.csv',
format='com.databricks.spark.csv',
header='true',
inferSchema='true')
```

```python
evaluator =  MulticlassClassificationEvaluator(labelCol='label', predictionCol='prediction', metricName='accuracy')
accuracy = evaluator.evaluate(predictions)
```

```python
print("Accuracy = %g "%(accuracy))
```

```python
predictions.rdd.map(tuple).take(2)
```

```python
metrics = MulticlassMetrics(predictions.rdd.map(tuple))
```

```python
metrics.confusionMatrix().toArray().transpose()
```

To get the error rate for the decision tree model, use the following code:

```python
print ("Error = %g " % (1.0 - accuracy))
```

To print out the accuracy as a percentage, use the following code:

```python
print ("Accuracy = %.2g" % (accuracy * 100))
```

In the last line of code, the confusion matrix is printed out. If the “transpose()” is
removed, the confusion matrix will be displayed as:

```python
metrics.confusionMatrix().toArray().transpose()
```
